```python
# netflix

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('netflix_titles.csv')
df.head(2)
```

```python
df.info()
```

```python
# change the date_added type to date_time
df['date_added']=pd.to_datetime(df['date_added'], format='mixed')
df.info()
```

```python
df.head(2)
```

```python
# create a new column (year)
df=df.dropna(subset=['year'])  #temporary drop
df['year']=df['date_added'].dt.year
df['year']=df['year'].astype('int64')
```

```python
df.head(2)
```

```python
df.info()
```

```python
# create a new column (year)

df['year']=df['date_added'].dt.year
```

```python
df.head(5)
```

```python
mask_movies = df["type"].eq("Movie") & df["duration"].str.contains("min", na=False)
mask_movies


```

```python
df.loc[mask_movies, "duration_min"] = (
    df.loc[mask_movies, "duration"].str.replace(" min", "", regex=False).astype('int64')
    )
```

```python
df['duration_min']=df['duration_min'].astype('int64')
```

```python
df.head(5)
```

```python
# explode genres
df['genre']=df['listed_in'].str.split(',')
df
```

```python
df.info()
```

```python
df_exploded=df.explode('genre')
df_exploded.head()
```

```python
df_exploded.info()
```

```python
trend=df.groupby('year')['year'].count().sort_values(ascending=False)
trend.head()
```

```python
trend.head()
```

```python
trend=pd.DataFrame(trend)
trend.info()
```

```python
plt.bar(trend.index, trend['year'] )
```

```python
sns.lineplot(data=trend, x=trend.index, y='year')
```

I discovered that the highest number of added items to Netflix network was (2016)element on year 2019.

```python
ratings=df.groupby('rating')['rating'].count().sort_values(ascending=False)
ratings=pd.DataFrame(ratings)
```

```python
ratings
```

```python
sns.barplot(data=ratings,x=ratings.index, y='rating')
plt.xlabel('Rating')
plt.xticks(rotation=90)   # to rotate the x-axis ticks by 45 degrees
plt.ylabel('Number of movies')

#plt.yticks(rotation=45)
plt.show()
```

```python
df.head(30)
```

```python
df['countries']=df['country'].str.split(',')
df.info()
```

```python
df_C=df.explode('countries')
df_C.head(30)
```

```python
df_C.info()
```

```python
netflix_country=df_C.groupby('countries')['countries'].count().sort_values(ascending=False)
netflix_country=pd.DataFrame(netflix_country)
netflix_country.head(5)
```

```python
plt.figure(figsize=(30,15))
sns.lineplot(data=netflix_country,x=netflix_country.index, y='countries')

plt.xlabel('Countries')
plt.xticks(rotation=90, fontsize=12)   # to rotate the x-axis ticks by 45 degrees

plt.show()
```

```python
# draw the 10 highest countries
sns.barplot(data=netflix_country.head(10),x=netflix_country.head(10).index, y='countries')
plt.xlabel('Countries')
plt.xticks(rotation=45)
plt.show()
```

The plot shows that USA (3205 items) has the highest contribution to netflix network, followed by Inda, then UK.

```python
# plot movies with duration

movies=df[df['type'].eq('Movie') & df['duration_min'].notna()].copy()
movies.sort_values('duration_min', ascending=False, inplace=True)
movies.head(5)
```

```python
sns.barplot(data=movies.head(10), x='duration_min', y='title')
```

```python
# plot histogram of movies with duration_min
sns.histplot(data=movies,x='duration_min', bins=30, kde=True)
```

From the histogram, we can say that the most common duration of movies is 90 min. The histogram shows a right-skewed distribution.
