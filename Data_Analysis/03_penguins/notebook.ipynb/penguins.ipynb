This dataset contains measurements for three penguin species observed in Antarctica.
Columns:

island: Location of study (Biscoe, Dream, Torgersen)

1. Load & Explore Data

Load dataset using sns.load_dataset("penguins").

Show first 10 rows, shape, and data types.

Check for missing values and handle them.

2. Univariate Analysis

Plot histogram of body_mass_g.

Countplot of species and island.

Check distribution of bill_length_mm with KDE.

3. Bivariate Analysis

Scatterplot: bill_length_mm vs bill_depth_mm, colored by species.

Boxplot: body_mass_g by species.

Violinplot: flipper_length_mm by island.

4. Multivariate Analysis

Pairplot of numerical features, colored by species.

Group by species and compute average body mass and flipper length.

5. Correlation Analysis

Compute correlation matrix of numeric features.

Plot heatmap using seaborn.

6. Insights (to answer)

Which species has the highest average body mass?

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load dataset
df = sns.load_dataset("penguins")

# 1. Basic exploration
print(df.head())
print(df.info())
print(df.describe())

# Check missing values
print(df.isnull().sum())

# 2. Histogram - Body Mass
sns.histplot(df["body_mass_g"], kde=True)
plt.title("Distribution of Penguin Body Mass")
plt.show()

# 3. Countplot - Species
sns.countplot(data=df, x="species")
plt.title("Count of Penguins per Species")
plt.show()

# 4. Scatterplot - Bill Length vs Bill Depth
sns.scatterplot(data=df, x="bill_length_mm", y="bill_depth_mm", hue="species")
plt.title("Bill Dimensions by Species")
plt.show()

# 5. Boxplot - Body Mass by Species
sns.boxplot(data=df, x="species", y="body_mass_g")
plt.title("Body Mass Distribution by Species")
plt.show()

# 6. Violinplot - Flipper Length by Island
sns.violinplot(data=df, x="island", y="flipper_length_mm", hue="species", split=True)
plt.title("Flipper Length by Island and Species")
plt.show()

# 7. Pairplot
sns.pairplot(df, hue="species")
plt.show()

# 8. Correlation Heatmap
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```
